```python
#data1.index = pd.to_datetime(data1.index)
#data1.index
#Stores list of all element in a variable called selected
selected = ['RPQ', 'BKK', 'ZWY', 'SDT', 'TJN', 'VPG', 'BQV','SJE']

sns.set_style('darkgrid')
counter = 1
fig = plt.figure(figsize=(8, 8))
fig.suptitle('Visitors correlations via date ',
    fontsize=14, position=(0.5, 1.0))
for i, name_i in enumerate(selected):
    for j in range(i + 1, len(selected)):
        name_j = selected[j]
        sub = fig.add_subplot(7, 5, counter)
        sub.set_title(name_i + ' vs ' + name_j,
            fontsize=10)
        sub.scatter(data1[name_i], data1[name_j], s=0.5)
        counter += 1
plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.tight_layout()
plt.show()
```

#### Interaction plot using hvplot

```python
selected_hvplot = ['RPQ', 'BKK', 'ZWY', 'SDT', 'TJN', 'VPG', 'BQV','SJE','XQE','YKT','WBK','TVJ','XWO','CCX','BWF','TPY',
                  'ZQL','WQW','YBS','WKL','YCI','UVQ','UDD','ACQ','VJV','CWN','XEZ','AKA','WVA','JJQ']
```

```python
plot = data1[selected_hvplot].hvplot.hist(
    frame_height=500, frame_width=500,
    xlabel='volume of visitors per week ', ylabel='Frequency/Counts',
    title='High Volume Visitors',
    alpha=0.5, muted_alpha=0, muted_fill_alpha=0, muted_line_alpha=0,
    tools=['pan', 'box_zoom', 'wheel_zoom', 'undo', 'redo', 'hover', 'save', 'reset']
)
hvplot.show(plot)
```

```python
plot2 = data1[selected_hvplot].sum().hvplot.bar(
    frame_height=200, frame_width=1000,
    xlabel='Unique Vistors Id', ylabel='Total counts of Visitors',
    title='Visitors Chart',
)
hvplot.show(plot2)
```
